## Simple Gen AI app using Langchain

```python
import os 
from dotenv import load_dotenv
load_dotenv()


os.environ['OPENAI_API_KEY']=os.getenv("OPENAI_API_KEY")
## LAngsmith Tracking
os.environ['LANGCHAIN_API_KEY']=os.getenv("LANGCHAIN_API_KEY")
os.environ['LANGCHAIN_TRACING_V2']="true"
os.environ['LANGCHAIN_PROJECT']=os.getenv("LANGCHAIN_PROJECT")
```

```python
## web scrapping
from langchain_community.document_loaders import WebBaseLoader
url="https://www.geeksforgeeks.org/artificial-intelligence/what-is-generative-ai/?"
```

```python
loader = WebBaseLoader(url)
docs=loader.load()
```

```python
print(docs)
```

```python
## load data ---> Docs----> Divide our text into chunks----->vector embedding------vecotr store db

from langchain_text_splitters import RecursiveCharacterTextSplitter
splitter=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)
splitter.split_documents(docs)

```

```python
from langchain_openai import OpenAIEmbeddings
embeddings=OpenAIEmbeddings()
```

```python
from langchain_community.vectorstores import FAISS
db=FAISS.from_documents(docs,embeddings)
```

```python
db
```

```python
from langchain_openai import ChatOpenAI
llm=ChatOpenAI(model="gpt-4o")
```

```python
query="Generative AI is a type of artificial intelligence designed to"
rst=db.similarity_search(query)
print(rst[0].page_content)
```

```python
# Retreival chain, Document Chain


from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_template(
    """
        Answer the following questions based on the provided context:
        <context>
        {context}
        </context>

        Question: {input}

"""
)

doc_chain=create_stuff_documents_chain(llm,prompt)
doc_chain
```

## Document Chain
 A chain that takes your documents and a prompt, stuffs the docs into the prompt, and asks the LLM to answer.

```python
from langchain_core.documents import Document

doc_chain.invoke({
    "input":"A Trace contains the series of steps that your application takes",
    "context":[Document(page_content="A Trace contains the series of steps that your application takes - from recieving an input, through intermediate steps, and generating a final output. Each of the individual steps is represented by a Run. A Project is a collection of traces")]
})
```

```python
## input------->Retriever------->vector store db

db
```

## Retriever chain
A chain that adds a retriever in front of the document chain.
Retriever = finds the most relevant documents from your vector database.
Then those docs are passed into the doc_chain.

```python
retriever=db.as_retriever()
from langchain.chains import create_retrieval_chain
ret_chain=create_retrieval_chain(retriever,doc_chain )
ret_chain
```

```python
response=ret_chain.invoke({"input":"What is generative AI?"})
response['answer']
```
